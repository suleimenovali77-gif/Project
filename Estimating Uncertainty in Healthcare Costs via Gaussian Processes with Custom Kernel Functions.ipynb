```python
# прописать алгоритм Гаусовского процесса
import numpy as np

# Целевая переменная (charges) - превращаем в вектор-столбец
# Не забываем про log-трансформацию, о которой писали в Methodology!
y = np.log(dataset['charges'].values).reshape(-1, 1)

# Признаки (X) - выбираем нужные колонки
features = ['age', 'sex', 'bmi', 'children', 'smoker', 'region']
X = dataset[features].values
```

```python
import numpy as np
import pandas as pd



# 2. Выбираем признаки (features)
features = ['age', 'sex', 'bmi', 'children', 'smoker', 'region']

# 3. СОЗДАЕМ X (вот этот шаг вы, скорее всего, пропустили или не запустили)
X = dataset[features].values 

# 4. Создаем y
y = np.log(dataset['charges'].values).reshape(-1, 1)

# Теперь print(X) сработает
print(X)
```

```python
# Создаем массив (используем круглые скобки и двойные квадратные для 2D-формы)
X_new = np.array([[39, 0, 30.4, 2, 0, 2]]) 

# Проверяем размерность
print(X_new.shape) # Должно быть (1, 6), так как у вас 6 признаков
```

```python
def custom_kernel(x1, x2, l=1.0, sigma_f=1.0):
    # Вычисляем квадрат евклидова расстояния
    sq_dist = np.sum(x1**2, axis=1).reshape(-1, 1) + np.sum(x2**2, axis=1) - 2 * np.dot(x1, x2.T)
    
    # Ваша формула: sin(exp(-dist / 2l^2))
    return sigma_f**2 * np.sin(np.exp(-sq_dist / (2 * l**2)))

# Вычисляем ковариацию между обучающими данными и новой точкой
k_star = custom_kernel(X, X_new)
print(k_star.shape) # Должно быть (n_samples, 1)
```

```python
# Параметры модели (гиперпараметры)
l = 1.0          # Длина корреляции
sigma_f = 1.0    # Амплитуда
sigma_n = 0.1    # Шум (важен для стабильности матрицы)

# 1. Вычисляем матрицу K (всех со всеми в обучающей выборке)
K = custom_kernel(X, X)

# 2. Добавляем шум к диагонали и инвертируем
# Вместо прямой инверсии используем решение системы уравнений (это быстрее и точнее)
Ky = K + sigma_n**2 * np.eye(len(X))

# 3. Вычисляем прогноз для X_new
# Используем формулу: k_*^T * (Ky^-1) * y
# В коде это эффективнее делать через np.linalg.solve
weights = np.linalg.solve(Ky, y)
f_star_log = np.dot(k_star.T, weights)

# 4. Возвращаем результат из логарифмического масштаба в доллары
predicted_charges = np.exp(f_star_log)

print(f"Предсказанная стоимость страховки для X_new: ${predicted_charges[0][0]:.2f}")
```

```python
import matplotlib.pyplot as plt

# 1. Создаем тестовую последовательность возрастов от 18 до 65 лет
age_range = np.linspace(0, 1, 100) # Предполагаем, что данные отнормированы
X_test = np.tile(X_new, (100, 1))   # Берем параметры X_new
X_test[:, 0] = age_range           # Меняем только колонку возраста

# 2. Считаем прогноз для всей линии
k_test = custom_kernel(X, X_test, l=l, sigma_f=sigma_f)
f_star_mean = np.dot(k_test.T, weights)

# 3. Считаем дисперсию (uncertainty)
# Для упрощения возьмем диагональ
v_inv = np.linalg.solve(Ky, k_test)
variance = sigma_f**2 - np.sum(k_test * v_inv, axis=0)
std_dev = np.sqrt(np.maximum(variance, 0))

# 4. Рисуем
plt.figure(figsize=(10, 6))
plt.fill_between(age_range * 50 + 18, 
                 np.exp(f_star_mean.flatten() - 1.96 * std_dev), 
                 np.exp(f_star_mean.flatten() + 1.96 * std_dev), 
                 alpha=0.2, color='blue', label='95% Confidence Interval')
plt.plot(age_range * 50 + 18, np.exp(f_star_mean), color='blue', lw=2, label='GP Prediction')
plt.scatter(X[:, 0] * 50 + 18, np.exp(y), alpha=0.1, color='black', s=10, label='Training Data')
plt.title("Insurance Charges Prediction with GP Uncertainty")
plt.xlabel("Age")
plt.ylabel("Charges ($)")
plt.legend()
plt.show()
```
